```python
#Divide by original dataset into test and train where test data contains 50 images each and train data contains 150 images each 

import os
import shutil
import random

def split_dataset(input_dir, train_dir, test_dir, train_size=56, test_size=14):
    # Ensure the train and test directories exist
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)
    
    # Loop through each subfolder in the input directory
    for subfolder in os.listdir(input_dir):
        subfolder_path = os.path.join(input_dir, subfolder)
        
        # Only process directories
        if os.path.isdir(subfolder_path):
            images = os.listdir(subfolder_path)
            random.shuffle(images)  # Shuffle to randomize the order
            
            # Split images into training and testing sets
            train_images = images[:train_size]
            test_images = images[train_size:train_size + test_size]
            
            # Create subfolders in the train and test directories
            train_subfolder_path = os.path.join(train_dir, subfolder)
            test_subfolder_path = os.path.join(test_dir, subfolder)
            os.makedirs(train_subfolder_path, exist_ok=True)
            os.makedirs(test_subfolder_path, exist_ok=True)
            
            # Copy the images to the train and test folders
            for img in train_images:
                shutil.copy(os.path.join(subfolder_path, img), os.path.join(train_subfolder_path, img))
            
            for img in test_images:
                shutil.copy(os.path.join(subfolder_path, img), os.path.join(test_subfolder_path, img))
                
            print(f"Processed {subfolder} - Train: {len(train_images)}, Test: {len(test_images)}")

# Define paths
input_directory = r"E:\MAJOR PROJECT\Dataset\Massey"   # Path to the folder containing the 16 subfolders
train_directory = r"E:\MAJOR PROJECT\Dataset\Massey_split\train"  # Path to save training data
test_directory = r"E:\MAJOR PROJECT\Dataset\Massey_split\test"    # Path to save testing data

# Run the function
split_dataset(input_directory, train_directory, test_directory)
```

```python
import os

def count_folders_images(directory):
    total_folders = 0
    total_images = 0
    folder_details = {}
    
    for root, dirs, files in os.walk(directory):
        num_folders = len(dirs)
        num_images = len([f for f in files if f.lower().endswith(('png', 'jpg', 'jpeg', 'bmp', 'gif'))])
        
        folder_details[root] = {'subfolders': num_folders, 'images': num_images}
        total_folders += num_folders
        total_images += num_images
    
    return total_folders, total_images, folder_details

# Example usage
directory = r"E:\MAJOR PROJECT\Dataset\Massey_split\train"  # Change this to your target folder
folders, images, details = count_folders_images(directory)
print(f"Total Folders: {folders}")
print(f"Total Images: {images}")
print("Details:")
for folder, info in details.items():
    print(f"{folder}: {info}")
```
